```python
import torch
from transformers import AutoProcessor, BarkModel

processor = AutoProcessor.from_pretrained("suno/bark")
model = BarkModel.from_pretrained("suno/bark")

# Check if CUDA is available and if so, move the model to GPU
if torch.cuda.is_available():
    model = model.to('cuda')

voice_preset = "v2/en_speaker_6"

inputs = processor("Once upon a time. There was a cat.  Really Ill parse huge chunks of text here.  Of course the voice has to change too.", voice_preset=voice_preset)

# If CUDA is available, also move the inputs to GPU
if torch.cuda.is_available():
    inputs = {name: tensor.to('cuda') for name, tensor in inputs.items()}

audio_array = model.generate(**inputs)

# Move the audio array back to CPU after generation
audio_array = audio_array.cpu().numpy().squeeze()
```

```python
from IPython.display import Audio

sample_rate = model.generation_config.sample_rate
Audio(audio_array, rate=sample_rate)
```
